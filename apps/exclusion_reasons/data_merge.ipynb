```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import re
```

```python
usage = pd.read_csv('data/internet_use.csv')
usage = usage.set_index(['Country Name','Country Code','Indicator Name','Indicator Code']).transpose()
usage = usage.stack().stack().stack().stack()
usage = usage.reset_index()
usage.columns = ['period','Indicator Code','Indicator Name','ISO','country','Individuals using the Internet (% of population)']
usage.drop(['Indicator Code','Indicator Name'],axis=1,inplace=True)
usage['period'] = pd.to_numeric(usage['period'], errors='coerce')
usage.head()
```

```python
barriers = pd.read_csv('data/internet_barriers.csv')
barriers.drop(['seriesID','seriesCode','seriesParent','seriesUnits','dataNote','dataSource','entityID'],axis=1,inplace=True)
barriers['cleanDescription'] = barriers['seriesDescription'].apply(lambda x: re.sub('Reasons for not having Internet access \\(i.e. Barriers\\): ','',x))
barriers[barriers['seriesDescription'] != barriers['seriesDescription'][0]]['seriesName'].value_counts()

barriers_1 = barriers[barriers['seriesDescription'] == barriers['seriesDescription'][0]]
barriers_1['Subregion'] = 'Total'
barriers_1.drop(['seriesDescription','cleanDescription','Subregion'],axis=1,inplace=True)
barriers_1 = barriers_1.set_index(['seriesName','entityIso','entityName','dataYear']).unstack(0)
barriers_1 = barriers_1.droplevel(0,axis=1)
barriers_1.reset_index(inplace=True)
barriers_1.columns = ['ISO', 'country', 'period',
       'Cost of service is too high', 'Cost of the equipment is too high',
       'Cultural reasons', 'Do not need the Internet',
       'Have access to the Internet elsewhere',
       'Internet service is available but it does not correspond to household needs',
       'Internet service is not available in the area',
       'No electricity in the household', 'Other reason',
       'Privacy or security concerns']
barriers_1.head()
```

```python
set(usage['country'].value_counts().index) - set(barriers_1['country'].value_counts().index)
```

```python
set(barriers_1['country'].value_counts().index) - set(usage['country'].value_counts().index)
```

```python
merged_df = barriers_1.merge(usage, how='left', left_on=['ISO','period'],right_on=['ISO','period'])
```

```python
import itertools

countries = set(merged_df['ISO'].values)
years = set(merged_df['period'].values)
dense_df = merged_df.copy()
dense_df.set_index(['ISO','period'],inplace=True)

new_index = set(itertools.product(countries,years)) - set(dense_df.index)

dense_df = pd.concat([merged_df.set_index(['ISO','period']), pd.DataFrame(index=pd.Index(new_index))])
```

```python
dense_df.reset_index(names=['ISO','period'],inplace=True)
dense_df.sort_values(['ISO','period'],inplace=True)
dense_df.head()
```

```python
wireless_df = pd.read_csv('./data/wireless_data_access.csv')
wireless_df = wireless_df[['GeoAreaName','TimePeriod','Value','SeriesCode']]
wireless_df = wireless_df[wireless_df['SeriesCode'] == 'IT_MOB_2GNTWK']
```

```python
name_map = {'United Kingdom of Great Britain and Northern Ireland':'United Kingdom',
 'Republic of Korea':'Korea (Rep. of)',
 'China, Macao Special Administrative Region':'Macao, China',
 'Czechia': 'Czech Republic',
 'China, Hong Kong Special Administrative Region': 'Hong Kong, China',
 
 }

wireless_df['GeoAreaName'] = wireless_df['GeoAreaName'].apply(lambda x: name_map[x] if x in name_map.keys() else x)
wireless_df.drop('SeriesCode',inplace=True,axis=1)
wireless_df.columns = ['country','period','2G Wireless Access']
```

```python
dense_df = dense_df.merge(wireless_df,left_on=['country_x','period'],right_on=['country','period'],how='left')
dense_df.drop(['country_y','country'],inplace=True,axis=1)
```

```python
cols = ['Cost of service is too high',
       'Cost of the equipment is too high', 'Cultural reasons',
       'Do not need the Internet', 'Have access to the Internet elsewhere',
       'Internet service is available but it does not correspond to household needs',
       'Internet service is not available in the area',
       'No electricity in the household', 'Other reason',
       'Privacy or security concerns',
       'Individuals using the Internet (% of population)',
       'country_x',
       '2G Wireless Access']

for col in cols:
    dense_df[col] = dense_df.groupby('ISO')[col].ffill()
```

```python
import geopandas as gpd
def get_geometries(users):
    url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
    gdf_ne = gpd.read_file(url)  # zipped shapefile
    gdf_ne = gdf_ne[["NAME", "CONTINENT", "POP_EST", "ISO_A3", 'geometry']]
    df = gdf_ne

    df.info()
    #  ['Country or Area', 'Internet Users', 'Population', 'Rank', 'Percentage', 'Rank.1']

    df1 = df[['ISO_A3', 'geometry']]
    df1.rename(columns={'ISO_A3': 'ISO'}, inplace=True)

    users_2 = users.copy()

    merged = users_2.merge(df1, how='right', on='ISO')

    return merged
```

```python
import dash
import dash_bootstrap_components as dbc
from dash import Input, Output, State, dcc, html, no_update
import plotly.express as px

df = merged_df


fig = px.choropleth(df, locations="ISO",
    color="Individuals using the Internet (% of population)", 
    hover_name="country_x", 
    projection="natural earth",
    animation_frame="period", animation_group="ISO",
    hover_data=["Cost of service is too high",'Cost of the equipment is too high','Cultural reasons'])
fig.update_traces(
    hoverinfo="none",
    hovertemplate=None,
)

# connect dash page to main app
app = dash.Dash(__name__, title='Dashboard', external_stylesheets=[dbc.themes.BOOTSTRAP])
app.layout = html.Div([
    dbc.Row([
        dbc.Col([
            dcc.Graph(id='scatter_geo', figure=fig)
        ]),
        dcc.Tooltip(id="graph-tooltip", direction='bottom')
    ])
])

@app.callback(
    Output("graph-tooltip", "show"),
    Output("graph-tooltip", "bbox"),
    Output("graph-tooltip", "children"),
    Input('scatter_geo','hoverData'))

def update_graph(hoverData):
    if hoverData is None:
        return False, no_update, no_update

    else:
        pt = hoverData["points"][0]['hovertext']
        pt2 = hoverData["points"][0]
        bbox = pt2["bbox"]

        dff = df[df['country_x']==pt]
        dff2 = dff.melt(id_vars=['country_x','ISO'], value_vars=cols)
        fig2 = px.pie(dff2, values='value', names='variable')
        fig2.update_layout(margin=dict(l=0, r=0, t=50, b=0), showlegend=False, title_text=pt)
        children = [
        html.Div([
            dcc.Graph(
                figure=fig2,style={"width": "200px", 'height':'200px'})
            ])
        ]

        return True, bbox, children

if __name__ == "__main__":
    app.run_server(debug=False)
```
